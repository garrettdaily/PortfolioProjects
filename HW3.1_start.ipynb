**4- Get a sample of the absenteeism dataset that has 50% of the original data with no replacement. Focusing on the "Absenteeism time in hours" variable, investigate whether your sample adequately represents the original data. Create at least THREE data summaries and THREE visualizations. It is up to you to decide how you want to investigate the representativeness of your sample. Make sure that you explain your findings as a comment or markdown.
```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Absenteeism_at_work.csv")
df_sample = df.sample(frac=0.5, replace=False, random_state=144)
```

```python
df_sample["Absenteeism time in hours"].mean()
```

```python
fig, ax1 = plt.subplots(figsize=(5, 3))
g1 = sns.kdeplot(data=df, x="Absenteeism time in hours", ax=ax1, label="Original")
g2 = sns.kdeplot(data=df_sample, x="Absenteeism time in hours", linestyle="--", ax=ax1, label="Sample")
plt.legend()
plt.show()
```

```python
df["source"] = "Original"
df_sample["source"] = "Sample"
df2 = pd.concat([df, df_sample], axis=0)
```

```python
sns.catplot(data=df2, x="Seasons", y="Absenteeism time in hours", kind="bar", hue="source");
```

```python
sns.catplot(data=df2, x="Seasons", y="Absenteeism time in hours", kind="point", hue="source");
```
